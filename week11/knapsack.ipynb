# Knapsack 0/1


```python
w = [None, 2, 1, 2, 3]
v = [None, 6, 4, 7, 8]
C_max = 6
```

```python
def optimal_value(w, v, C_max):
    n = len(w) - 1
    # Subproblem table
    S = [[0] * (C_max + 1) for _ in range(n + 1)]
    for i in range(1, n + 1):
        for r in range(1, C_max + 1):
            if w[i] > r:
                S[i][r] = S[i - 1][r]
            else:
                S[i][r] = max(S[i - 1][r], v[i] + S[i - 1][r - w[i]])
    return S
```

```python
def trace_back(S, w, v, C_max):
    # Start from bottom, right
    included_items = []
    i = len(w) - 1
    r = C_max
    while i > 0 and r >= 0:
        if S[i][r] != S[i - 1][r]:
            included_items.append(i)
            r -= w[i]
        i -= 1
    return included_items
```

```python
S = optimal_value(w, v, C_max)
print(f"Optimal value is ${S[-1][-1]}m")
for row in S:
    print(row)
included_items = trace_back(S, w, v, C_max)
print(f"Here's what to take away: {included_items}")
```
