# Knapsack 0/1


```python
# Weights and values of the items.
# The first element is None to make the indexing start from 1.
w = [None, 2, 1, 2, 3]
v = [None, 6, 4, 7, 8]
C_max = 6
```

```python
from __future__ import annotations
from typing import Union


def optimal_value(w: list[int], v: list[int], C_max: int) -> list[list[int]]:
    """Returns the optimal value and the table of subproblems."""
    # Shortcut to the number of items.
    n = len(w) - 1
    # Initialize the table of subproblems. This ensures that base cases
    # at S(i,0) and S(0,r) are all 0.
    S = [[0] * (C_max + 1) for _ in range(n + 1)]
    # Compute all subproblems in increasing order of i and r.
    for i in range(1, n + 1):
        # Consider subproblems among the first i artifacts.
        for r in range(C_max + 1):
            # Consider the subproblem with residual capacity r.
            if w[i] > r:
                # The i-th artifact cannot be included in the optimal solution,
                # because of weight restrictions, so the optimal value is the same
                # as the optimal value without the i-th artifact.
                S[i][r] = S[i - 1][r]
            else:
                # The i-th artifact can be included in the optimal solution,
                # so the optimal value is the maximum of the optimal value
                # without the i-th artifact and the optimal value with
                # the i-th artifact.
                S[i][r] = max(S[i - 1][r], S[i - 1][r - w[i]] + v[i])
    # Done! Return the entire subproblem table and let users figure out
    # the optimal value from it.
    return S
```

```python
def trace_back(S: list[list[int]], w: list[int], v: list[int], C_max: int) -> list[int]:
    """Returns the list of items included in the optimal solution."""
    # List of items included in the optimal solution.
    # We will build this list in reverse order.
    included_items = []
    # Start from the last item and from the full capacity.
    i = len(w) - 1
    r = C_max
    while i > 0 and r >= 0:
        if S[i][r] != S[i - 1][r]:
            # The i-th artifact is included in the optimal solution.
            included_items.append(i)
            r -= w[i]
        i -= 1
    return included_items
```

```python
# Demonstration of the optimal_value function.
S = optimal_value(w, v, C_max)
print("Optimal value:", S[-1][-1])
print("Table of subproblems:")
for row in S:
    print(row)
included_items = trace_back(S, w, v, C_max)
print("Included items:", included_items)
```
