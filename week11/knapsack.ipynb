# Knapsack 0/1



```python
# A simple data set for weights and values of four items.
# The first element of each list is None to make the indexing
# start at 1 for convenience.
w = [None, 2, 1, 2, 3]
v = [None, 6, 4, 7, 8]
C_max = 6
```

```python
def optimal_value(w, v, C_max):
    """Returns the subproblem table S for the 0-1 knapsack problem."""
    # Number of items
    n = len(w) - 1
    # Initialize the subproblem table S with zeros. This takes care of
    # the base cases where either no items are considered or the capacity is zero.
    S = [[0] * (C_max + 1) for _ in range(n + 1)]
    # Fill the subproblem table S using the dynamic programming approach.
    for i in range(1, n + 1):
        for r in range(1, C_max + 1):
            # If the weight of the current item exceeds the current capacity,
            # we cannot include it.
            if w[i] > r:
                # We cannot include the current item, so the optimal value is
                # the same as the optimal value for the previous items with the
                # same capacity.
                S[i][r] = S[i - 1][r]
            else:
                # The weight of the current item is less than or equal to
                # the current capacity.We have the choice to include the current item or not.
                # We take the maximum.
                # In the expression below, r-w[i] is the remaining capacity
                # allowing room for the current item, and S[i - 1][r - w[i]]
                # is the optimal value for that remaining capacity before we
                # include the current item.
                S[i][r] = max(S[i - 1][r], v[i] + S[i - 1][r - w[i]])
                #             ===========  =========================
                #             Leave the    Take the current item
                #             current
                #             item
    return S
```

```python
def trace_back(S, w, v, C_max):
    """Returns the list of included items in the optimal solution."""
    # Indices to start from the bottom-right corner of the subproblem table S
    i = len(w) - 1
    r = C_max
    # List to store the indices of included items
    included_items = []
    # Trace back through the subproblem table S to find which items are included 
    # in the optimal solution.
    while i > 0 and r >= 0:
        # If the value in the current cell S[i][r] is different from the 
        # value in the cell above S[i - 1][r], it means that the current 
        # item i is included in the optimal solution.
        if S[i][r] != S[i - 1][r]:
            # Add current item i to the list of included items
            included_items.append(i)
            # Update the remaining capacity r by subtracting 
            # the weight of the included item
            r -= w[i]
        # Move to the previous item
        i -= 1
    return included_items
```

```python
# --- Demo ---
S = optimal_value(w, v, C_max)
print(f"Optimal value is ${S[-1][-1]}m")
for row in S:
    print(row)
included_items = trace_back(S, w, v, C_max)
print(f"Here's what to take away: {included_items}")
```
