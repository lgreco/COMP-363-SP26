```python
import random, math

def f(x):
    return math.sin(x)

def monte_carlo_integration(f, a, b, n_samples):

    total = 0
    for _ in range(n_samples):
        x = random.uniform(a,b)
        total = total + f(x)
    average_height = total /n_samples
    return (b-a)*average_height

print(monte_carlo_integration(f, 0, 3.14159, 5000))
```
