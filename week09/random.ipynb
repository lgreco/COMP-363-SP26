```python
from random import randint

def generate_random_list(size, lower_bound, upper_bound):
    return [randint(lower_bound, upper_bound) for _ in range(size)]

def find_value_las_vegas(lst, value):
    trials = 0
    while True:
        trials += 1
        index = randint(0, len(lst) - 1)
        if lst[index] == value:
            return index, trials

if __name__ == "__main__":
    size = 10_000
    lower_bound = 1
    upper_bound = 1000
    random_list = generate_random_list(size, lower_bound, upper_bound)
    print("Generated List:", random_list)

    value_to_find = random_list[randint(0, size - 1)]  # Ensure the value is in the list
    print("Value to find:", value_to_find)

    index_found, trials = find_value_las_vegas(random_list, value_to_find)
    print(f"Value {value_to_find} found at index: [{index_found}] in {trials} trials")
```
