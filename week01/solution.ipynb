# Assignment 01 solution: quantitative analysis of mergesort




So, empirically we've shown that for mergesort, $T(n) \approx n\log_2 n$ or (to use better notation) $T(n)\in\mathcal O(n\log_2 n)$. Now, of course we have to prove that this is the case for any size array, not for the special case $n=8$ we analyzed in class.

To prove that 

$$T(n) = 2T(n/2) + f(n)$$

for any $n$ that is a power of 2, i.e., $n=2^p$ with $p\in\mathbb N$ we start by writing

$$
\begin{align*}
T(n) &= 2T\!\left(\frac{n}{2}\right) + f(n) \\[4pt]
     &= 2\left[ 2T\!\left(\frac{n}{4}\right) + f\!\left(\frac{n}{2}\right) \right] + f(n) \\[4pt]
     &= 2\left[ 2\left[ 2T\!\left(\frac{n}{8}\right) + f\!\left(\frac{n}{4}\right) \right] + f\!\left(\frac{n}{2}\right) \right] + f(n) \\[4pt]
     &= 2\left[ 2\left[ 2\left[ 2T\!\left(\frac{n}{16}\right) + f\!\left(\frac{n}{8}\right) \right] + f\!\left(\frac{n}{4}\right) \right] + f\!\left(\frac{n}{2}\right) \right] + f(n)
\end{align*}
$$

Eventually we'll run out of terms -- like we did with the case of $n=8$ when the sum finally got the form $T(8) = 8T(1)+4f(2)+2f(4)+8f(1)$. In other words we'll run out of denominators to divide $n$. Noting that the fraction in $T\!\left(\cdot\right)$ progresses with denominators that are powers of two, we can rewriten $T(n)$ as:


$$
\begin{align*}
T(n)
&= 2^k\,T\!\left(\frac{n}{2^k}\right)
   + \sum_{i=0}^{k-1} 2^i\, f\!\left(\frac{n}{2^i}\right)
\end{align*}
$$

In the expression above we expect $\dfrac{n}{2^k}=1$ and therefore $k=\log_2n$. And remembering that, for mergesort, $T(1)=1$ and  $f(n)=n$, the expression can be written as:


$$
\begin{align*}
T(n)
&= 2^k\,T\!\left(\frac{n}{2^k}\right)
   + \sum_{i=0}^{k-1} 2^i\, f\!\left(\frac{n}{2^i}\right) \\
& = nT(1) + \sum_{i=0}^{k-1} 2^i \frac{n}{2^i} \\
& = n + \sum_{i=0}^{k-1} n \\
& = n + kn \\
& = n(1+k) \\
& = n(1+\log_2n) \approx n\log_2n
\end{align*} 
$$
